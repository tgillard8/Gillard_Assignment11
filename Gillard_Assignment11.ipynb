```python
# dictionaries with each suit and a value range from 2 to 10. each suit is a key to another dictionary. 
deck_suit = {
    "clubs": {str(value): {} for value in range(2, 11)},
    "diamonds": {str(value): {} for value in range(2, 11)},
    "hearts": {str(value): {} for value in range(2, 11)},
    "spades": {str(value): {} for value in range(2, 11)}
}

# face cards for each suit
for suit in deck_suit:
    deck_suit[suit].update({"Jack": {}, "Queen": {}, "King": {}, "Ace": {}})

#printing loop
for suit, cards in deck_suit.items():
    print(f"{suit}:")  # prints suit
    for card in cards:  # loop for cards
        print(f"  {card}")  # print card
    print()  # line between each suit
```

```python
#list for all card values including numbers and faces
values = list(map(str, range(2, 11))) + ["Jack", "Queen", "King", "Ace"]

#dictionary for each card value of which points to another dictionary with suits as keys
deck_value = {value: {suit: {} for suit in ["Clubs", "Diamonds", "Hearts", "Spades"]} for value in values}

#printing loop
for value, suits in deck_value.items():
    print(f"{value}:")  # print value of card
    for suit in suits:  # loop for each suit
        print(f"  {suit}")  # print suit
    print()  # line between card values
```

```python
def translate(rnasequence):
    # codon mapping table
    codon_table = {
        'UUU': 'F', 'CUU': 'L', 'AUU': 'I', 'GUU': 'V',
        'UUC': 'F', 'CUC': 'L', 'AUC': 'I', 'GUC': 'V',
        'UUA': 'L', 'CUA': 'L', 'AUA': 'I', 'GUA': 'V',
        'UUG': 'L', 'CUG': 'L', 'AUG': 'M', 'GUG': 'V',
        'UCU': 'S', 'CCU': 'P', 'ACU': 'T', 'GCU': 'A',
        'UCC': 'S', 'CCC': 'P', 'ACC': 'T', 'GCC': 'A',
        'UCA': 'S', 'CCA': 'P', 'ACA': 'T', 'GCA': 'A',
        'UCG': 'S', 'CCG': 'P', 'ACG': 'T', 'GCG': 'A',
        'UAU': 'Y', 'CAU': 'H', 'AAU': 'N', 'GAU': 'D',
        'UAC': 'Y', 'CAC': 'H', 'AAC': 'N', 'GAC': 'D',
        'UGU': 'C', 'CGU': 'R', 'AGU': 'S', 'GGU': 'G',
        'UGC': 'C', 'CGC': 'R', 'AGC': 'S', 'GGC': 'G',
        'UGG': 'W', 'CGG': 'R', 'AGG': 'R', 'GGG': 'G'
    }

    protein = []

    # process each codon within rna sequence
    for i in range(0, len(rnasequence), 3):
        codon = rnasequence[i:i+3]
        if len(codon) == 3:
            amino_acid = codon_table.get(codon, '')  #empty if not found
            if amino_acid:
                protein.append(amino_acid)

    #convert list of amino acids to string
    protein_str = ''.join(protein)

    #print length of protein
    print(f"Length of Protein: {len(protein)}")
    
    return protein_str

#input for user rna sequence
print("Enter RNA sequence:")
rna_seq = input().strip().upper()
protein = translate(rna_seq)
print("Encoded Protein:", protein)

```
